```python
import pandas as pd
# Charger le fichier de données
df = pd.read_excel("Election.xlsx")
df.head(1000)
```

```python
# Supprimer les colonnes inutiles
colonnes_a_supprimer = ["Code du département", "Code de la circonscription", "Code de la commune", "Code du b.vote"]
df = df.drop(columns=colonnes_a_supprimer)

# Renommer les colonnes
noms_nouvelles_colonnes = {
    "Libellé du département": "Département",
    "Libellé de la circonscription": "Circonscription",
    "Libellé de la commune": "Commune",
    "Inscrits": "Nombre Inscrits",
    "Abstentions": "Nombre Abstentions",
    "% Abs/Ins": "Pourcentage Abstentions/Inscrits",
    "% Vot/Ins": "Pourcentage Votes/Inscrits",
    "% Blancs/Ins": "Pourcentage Votes Blancs/Inscrits",
    "% Blancs/Vot": "Pourcentage Votes Blancs/Votés",
    "% Nuls/Ins": "Pourcentage Votes Nuls/Inscrits",
    "% Nuls/Vot": "Pourcentage Votes Nuls/Votés",
    "% Exp/Ins": "Pourcentage Votes Exprimés/Inscrits",
    "% Exp/Vot": "Pourcentage Votes Exprimés/Votés",
}
df = df.rename(columns=noms_nouvelles_colonnes)

# Convertir les types de données
df["Nombre Inscrits"] = df["Nombre Inscrits"].astype(int)
df["Nombre Abstentions"] = df["Nombre Abstentions"].astype(int)
df["Pourcentage Abstentions/Inscrits"] = df["Pourcentage Abstentions/Inscrits"].astype(float)
df["Pourcentage Votes/Inscrits"] = df["Pourcentage Votes/Inscrits"].astype(float)
df["Pourcentage Votes Blancs/Inscrits"] = df["Pourcentage Votes Blancs/Inscrits"].astype(float)
df["Pourcentage Votes Blancs/Votés"] = df["Pourcentage Votes Blancs/Votés"].astype(float)
df["Pourcentage Votes Nuls/Inscrits"] = df["Pourcentage Votes Nuls/Inscrits"].astype(float)
df["Pourcentage Votes Nuls/Votés"] = df["Pourcentage Votes Nuls/Votés"].astype(float)
df["Pourcentage Votes Exprimés/Inscrits"] = df["Pourcentage Votes Exprimés/Inscrits"].astype(float)
df["Pourcentage Votes Exprimés/Votés"] = df["Pourcentage Votes Exprimés/Votés"].astype(float)
# Ajouter d'autres conversions si nécessaire

# Traiter les valeurs manquantes
df.dropna(inplace=True)

# Afficher les premières lignes du DataFrame nettoyé
print(df.head())
# Exporter le DataFrame nettoyé vers un fichier Excel
df.to_excel("donnees_el_propres.xlsx", index=False)
```

```python
import matplotlib.pyplot as plt
#visualisation 
plt.figure(figsize=(10,6))
plt.bar(df['Département'], df['Nombre Abstentions'])
plt.xlabel('Département')
plt.ylabel('Nombre d\'abstentions')
plt.xticks(rotation=90)
plt.title('Abstention par département')
plt.show()
```

```python
labels = ['Exprimés', 'Blancs', 'Nuls']
sizes = [df['Pourcentage Votes Exprimés/Votés'].mean(), df['Pourcentage Votes Blancs/Votés'].mean(), df['Pourcentage Votes Nuls/Votés'].mean()]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Répartition des votes')
plt.show()
```

```python
# Charger les données de sécurité en spécifiant le délimiteur
df_securite = pd.read_csv("Sécutité.csv", delimiter=";")
# Afficher les premières lignes du DataFrame
df_securite.head(1000)
```

```python
# Supprimer les colonnes inutiles si nécessaire
colonnes_a_supprimer = ["classe", "annee"]  # Liste des colonnes à supprimer si elles ne sont pas pertinentes
df_securite_clean = df_securite.drop(columns=colonnes_a_supprimer)

# Traiter les valeurs manquantes si nécessaire
df_securite_clean.dropna(inplace=True)  # Supprimer les lignes avec des valeurs manquantes
# Gérer les doublons si nécessaire
df_securite_clean.drop_duplicates(inplace=True)

# Renommer les colonnes si nécessaire pour plus de clarté
nouvelles_noms = {
    "Code.région": "Code Région",
    "unité.de.compte": "Unité de Compte",
    "millPOP": "Population (milliers)",
    "millLOG": "Logarithme de Population (milliers)",
    "faits": "Nombre de Faits",
    "POP": "Population",
    "tauxpourmille": "Taux pour Mille",  # Renommer la colonne 'tauxpourmille'
    "LOG": "Logarithme de Population"   # Renommer la colonne 'LOG'
}
df_securite_clean = df_securite_clean.rename(columns=nouvelles_noms)

# Exporter le DataFrame nettoyé vers un fichier Excel
df_securite_clean.to_excel("donnees_s_propres.xlsx", index=False)
```

```python
plt.figure(figsize=(10,6))
plt.bar(df_securite_clean['Code Région'], df_securite_clean['Nombre de Faits'])
plt.xlabel('Région')
plt.ylabel('Nombre de faits')
plt.xticks(rotation=90)
plt.title('Nombre de faits par région')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df_securite_clean['Population'], df_securite_clean['Taux pour Mille'])
plt.xlabel('Population')
plt.ylabel('Taux pour mille')
plt.title('Population vs Taux pour mille')
plt.show()
```

```python
# Charger les données de economie en spécifiant le délimiteur
df_eco = pd.read_csv("economie.csv")
# Afficher les premières lignes du DataFrame
df_eco.head(1000)
```

```python
# Traiter les valeurs manquantes si nécessaire
df_eco.dropna(inplace=True)
# Gérer les doublons si nécessaire
df_eco.drop_duplicates(inplace=True)
# Supprimer les colonnes inutiles si nécessaire
colonnes_a_supprimer_economique = ["Libellé"]
df_economique_clean = df_eco.drop(columns=colonnes_a_supprimer_economique)
# Renommer les colonnes si nécessaire pour plus de clarté
nouvelles_noms = {
    "Nom": "Nom",
    "Longueur": "Longueur",
    "Type": "Type",
    "Ordre": "Ordre",
}
df_eco = df_eco.rename(columns=nouvelles_noms)

# Exporter le DataFrame nettoyé vers un fichier Excel
df_eco.to_excel("donnees_eco_propres.xlsx", index=False)
```

```python
plt.figure(figsize=(10,6))
df_economique_clean['Type'].value_counts().plot(kind='bar')
plt.xlabel('Type d\'activité')
plt.ylabel('Nombre d\'occurrences')
plt.title('Distribution des types d\'activités économiques')
plt.show()
```

```python
plt.figure(figsize=(8,8))
df_eco['Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Répartition des activités économiques')
plt.ylabel('')
plt.show()
```

```python
# Charger les données de chaque fichier
df_election = pd.read_excel("donnees_el_propres.xlsx")
df_securite = pd.read_excel("donnees_s_propres.xlsx")
df_economie = pd.read_excel("donnees_eco_propres.xlsx")

# Ajouter une colonne 'Source' pour identifier la provenance des données
df_election['Source'] = 'Election'
df_securite['Source'] = 'Securite'
df_economie['Source'] = 'Economie'

# Concaténer les DataFrames
df_combined = pd.concat([df_election, df_securite, df_economie], ignore_index=True)

# Afficher les premières lignes du DataFrame combiné
print(df_combined.head())

# Exporter le DataFrame combiné vers un fichier Excel
df_combined.to_excel("donnees_combinees.xlsx", index=False)
```
